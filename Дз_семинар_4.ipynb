##Задание 3

###Исследуйте, какие характеристики влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import seaborn as sns

plt.figure(figsize = (10, 6))
sns.boxplot(x = df['condition'], y = df['price'],  showfliers =False)

plt.title("Взаимосвязь стоимости дома и нсколько у него хорошее состояние")
plt.xlabel("condition")
plt.ylabel("price, евро");
```

**Состояние дома оценивается от 1 до 5. Видно, что средняя стоимость домов с показателем равным 1 и 2, 3 и 4 одинакова. Однако видна общая закономерность 1=2 -> 3=4 -> 5: чем лучше состояние дома, тем выше стоимость.**

```python
plt.figure(figsize = (8, 6))
sns.lineplot(x = df['yr_built'], y = df['price'])

plt.title("Взаимосвязь стоимости дома и года постройки")
plt.xlabel("yr_built")
plt.ylabel("price, евро");
```

**Самые дорогие дома построены до 1938 года, несмотря на почтенный возраст они сейчас ценятся не менее чем новостройки. Самые дешовые дома 1942-1945 годов. Скорее всего, на их строительстве в этот период экономили, в связи с экономическим и финансовыми трудностями, связанными с ВОВ.**

```python
plt.figure(figsize = (8, 6))
sns.scatterplot(x = df['sqft_basement'], y = df['price'])

plt.title("Взаимосвязь стоимости дома и размера подвала")
plt.xlabel("sqft_basement, кв.м")
plt.ylabel("price, евро");
```

**Наличие подвала не оказывает особенного вляния на стоимость дома**

```python
plt.figure(figsize = (8, 6))
sns.jointplot(x = df['grade'], y = df['price'], kind ='kde')

plt.title("Взаимосвязь стоимости дома и конструкции/дизайна")
plt.xlabel("grade")
plt.ylabel("price, евро");
```

**Grade- показатель конструкции/дизайна здания от 1 до 13**
**Видна выраженная зависимость. Чем интереснее конструкция/дизайн, чем дороже стоит дом. Особенно большое количество домов с показателем от 6 до 8.**

```python
plt.figure(figsize = (8, 6))
sns.barplot(x = df['bedrooms'], y = df['price'])

plt.title("Взаимосвязь стоимости дома и количества спален")
plt.xlabel("bedrooms")
plt.ylabel("price, евро");
```

**Максимальная стоимость у домов с 8 спальнями, минимальная с одной. Чем меньше(или больше) 8 комнат, стоимость идет вниз. Ярко выражена взаимосвязь в виде параболлы, ветви которой направлены вниз. Есть группы домов, выбивающихся из этой зависимости:**

#####1. С 33 спальнями. Скорее всего это дома с большой площадью, соответственно, и стоимость их выше.
#####2. Дома без спален. Скорее всего это не жилые дома, а дома , которые используются для офисов или других специализированных нужд.
