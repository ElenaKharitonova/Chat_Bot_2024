```python
import pandas as pd
df = pd.read_csv('/content/kc_house_data.csv', encoding = 'latin-1')
df.head()
```

##Домашнее задание 1

###1.1 Создать новый признак price_per_sq_lot, который будет содержать среднюю стоимость за один кв. метр общей площади


```python
df['price_per_sq_lot'] = df['price'] / df['sqft_lot15']
df.head()
```

###1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома

Если реновации дома не было, то в новом признаке поставьте 0

```python
def renovated(x):
  if x > 0:
    res = x
  else:
    res = 0
  return int(res)

df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df['delta_renovated'] = df['delta_renovated'].apply(renovated)
df.head()
```

###1.3 Создайте признаки года продажи, месяца продажи

```python
df['year_ch'] = pd.to_datetime(df['date'], format='%Y%m%dT000000').dt.year
df['month_ch'] = pd.to_datetime(df['date'],format='%Y%m%dT000000').dt.month
df.head()
```

###1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns = ['date', 'zipcode', 'lat', 'long'], inplace = True)
df.head()
```
